# Recommender

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
%matplotlib inline
column_names = ['user_id', 'item_id', 'rating', 'timestamp']
df = pd.read_csv('u.data', sep = '\t', names = column_names)
movies_titles = pd.read_csv('Movie_Id_Titles')
```

```python
df = pd.merge(df,movies_titles, on= 'item_id')
df.head()
```

```python
#df.groupby('title')['rating'].mean().sort_values(ascending=False).head(5)
#df.groupby('title')['rating'].mean().sort_values(ascending=False).sample(5)
#df.groupby('title')['rating'].count().sort_values(ascending=False).head(5)
```

```python
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
ratings['num_ratings'] = pd.DataFrame(df.groupby('title')['rating'].count())
ratings.head()
```

```python
movie_matrix = df.pivot_table(index = 'user_id', columns='title',values = 'rating')
movie_matrix.head()
```

```python
#star=movie_matrix['Star Wars (1977)']
#liar=movie_matrix['Liar Liar (1997)']
similar_to_star = movie_matrix.corrwith(movie_matrix['Star Wars (1977)'])
similar_to_liar = movie_matrix.corrwith(movie_matrix['Liar Liar (1997)'])
similar_to_star
```

```python
df_corr_starwars = pd.DataFrame(similar_to_star, columns = ['Correlation'])
df_corr_starwars.dropna(inplace = True)
df_corr_starwars.sort_values(by ='Correlation',ascending=False).head()
```

```python
df_corr_starwars = df_corr_starwars.join(ratings['num_ratings'])
df_corr_starwars.head()
```

```python
df_corr_starwars[df_corr_starwars['num_ratings']>100].sort_values(by='Correlation',ascending=False)
```
